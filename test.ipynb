```python
def renameColumns(df:pd.DataFrame) -> None:
    df.columns = df.columns.str.replace('[^a-zA-Z]', '', regex=True)
```

```python
renameColumns(df)
df.columns.tolist()
```

```python
df['DATEDENAISSANCE'] = pd.to_datetime(df['DATEDENAISSANCE'], errors='coerce')
df['DATEDEPRISEDESERVICE'] = pd.to_datetime(df['DATEDEPRISEDESERVICE'], errors='coerce')
```

```python
columns_mode = [
    'LIEUDENAISSANCE', 
    'SITUATIONSOCIOPROFESSIONNELlinscription', 
    'INTITULEPOSTE', 
    'STRUCTUREDIRECTIONPOLE', 
    'ENTREPRISES', 
    'TYPEDECONTRAT', 
    'STATUTACTUELenposteounon',
    'DATEDEPRISEDESERVICE',
    'DATEDENAISSANCE'

]

for col in columns_mode:
    if col in df.columns and not df[col].dropna().empty:  
        mode_value = df[col].mode().iloc[0]  
        df[col] = df[col].fillna(mode_value)  
    else:
        print(f"Impossible de calculer le mode pour la colonne {col} : colonne vide ou inexistante.")


columns_fixed = {
    'ADRESSE': 'Non spécifié',
    'CONTACTDURGENCE': 'Non fourni',
    'STATUTMATRIMONIALE': 'Non spécifié',
    'PROFILAGE': 'Non défini',
    'COMMENTAIRE': 'Aucun commentaire',
    'CONTACTENTREPRISE': 'Non précisé'
}

for col, value in columns_fixed.items():
    if col in df.columns:
        df[col] = df[col].fillna(value)  
    else:
        print(f"La colonne {col} n'existe pas.")


columns_mean = ['NIVEAUDETUDElinscription', 'REMUNERATION', 'DUREEMOIS']

for col in columns_mean:
    if col in df.columns and pd.api.types.is_numeric_dtype(df[col]):  
        mean_value = df[col].mean()
        df[col] = df[col].fillna(mean_value)  
    else:
        print(f"La colonne {col} n'est pas numérique ou n'existe pas.")


columns_zero = ['DUREEMOIS']
for col in columns_zero:
    if col in df.columns:
        df[col] = df[col].fillna(0)  
    else:
        print(f"La colonne {col} n'existe pas.")
```
